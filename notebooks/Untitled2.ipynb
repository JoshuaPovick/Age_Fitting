```python
inclins = np.linspace(0,180)

fig = plt.figure(figsize=[12,7])

ax = fig.add_subplot(111)

ax_top = ax.secondary_xaxis('top')
ax_top.set_xticks([25.86,34.7,63.3])
ax_top.set_xticklabels([r'CHOI18',r'VDM01',r'KADRMAS22'],rotation=45)

# ax.plot(recip_anchors,fitz_alebv(np.reciprocal(recip_anchors)),label='Fitzpatrick')
# ax.plot(recip_anchors,alebv_ccm,label='CCM')

lmcrad = 8

for i in np.arange(0.,3.0,0.5): #range(10)[::2]:
    ax.plot(inclins,2*lmcrad*np.sin(i*np.radians(inclins)),label='{}'.format(i))
    
ax.grid()

ax.legend(ncol=2)

ax.axvline(25.86,c='k',lw=3.0,ls='--')
ax.axvline(34.7,c='k',lw=3.0,ls='--')
ax.axvline(63.3,c='k',lw=3.0,ls='--')

ax.set_xlim(15,90)
# ax.set_ylim(-0.5,2.5)

ax.set_ylabel(r'2Rsin(n$\times \iota$)')
ax.set_xlabel(r'$\iota$');
```
