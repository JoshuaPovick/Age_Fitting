```python
#Astropy
import astropy
from astropy.io import fits,ascii
from astropy.table import Table

# Matplotlib
import matplotlib
import matplotlib.pyplot as plt
# %matplotlib inline
matplotlib.rcParams.update({'font.size': 25})

#Numpy/Scipy
import numpy as np
from scipy.interpolate import InterpolatedUnivariateSpline as IUS
from scipy.optimize import curve_fit
```

# Data

```python
kasc672 = fits.getdata('/Users/joshuapovick/Desktop/Research/APOKASC/APOKASC_cat_v6.7.2.fits.gz')
kasc3 = ascii.read('MLtesterr3bDR16A3PalphaM_july5.out')

# xmatch APOKASC catalogs

_, k672idx, k3idx = np.intersect1d(kasc672['2MASS_ID'],kasc3['2MASSID'],return_indices=True)
k672 = kasc672[k672idx]
k3 = kasc3[k3idx]

dr16 = fits.getdata('/Users/joshuapovick/Desktop/Research/fits/allStarLite-r12-l33.fits.gz')
dr17 = fits.getdata('/Users/joshuapovick/Desktop/Research/fits/allStarLite-dr17-synspec.fits.gz')
```

```python
print(len(k672['2MASS_ID'])==len(k3['2MASSID']))

print(np.where((k672['2MASS_ID']==k3['2MASSID'])==False))
```

# Xmatch DRs

```python
# xmatch drs
_, fdr16idx, fdr17idx = np.intersect1d(dr16['APOGEE_ID'],dr17['APOGEE_ID'],return_indices=True)

fdr16 = dr16[fdr16idx]
fdr17 = dr17[fdr17idx]

# xmatch with APOKASC

_, k3idxa, fdr16idxa = np.intersect1d(k3['2MASSID'],fdr17['APOGEE_ID'],return_indices=True)

fdr_k672 = k672[k3idxa]
fdr_k3 = k3[k3idxa]
fdr_dr16 = fdr16[fdr16idxa]
fdr_dr17 = fdr17[fdr16idxa]

ageevs = np.where((fdr_k3['IntAge']>-10.)&(fdr_k672['APOKASC3_CONS_EVSTATES']==1.0)&
                  (fdr_dr17['GAIAEDR3_PARALLAX']/fdr_dr17['GAIAEDR3_PARALLAX_ERROR']>3.0))

fdr_k672 = fdr_k672[ageevs]
fdr_k3 = fdr_k3[ageevs]
fdr_dr16 = fdr_dr16[ageevs]
fdr_dr17 = fdr_dr17[ageevs]
```

```python
plt.scatter(fdr_dr16['FE_H'],fdr_dr17['FE_H'])
plt.xlabel(r'ASPCAP [Fe/H] DR16')
plt.ylabel(r'ASPCAP [Fe/H] DR17')
plt.show()

plt.scatter(fdr_dr16['FE_H'],fdr_k3['Fe/H'])
plt.xlabel(r'ASPCAP [Fe/H] DR16')
plt.ylabel(r'APOKASC 3 [Fe/H]')
plt.show()

plt.scatter(fdr_dr17['FE_H'],fdr_k3['Fe/H'])
plt.xlabel(r'ASPCAP [Fe/H] DR17')
plt.ylabel(r'APOKASC 3 [Fe/H]')
plt.show()
```
